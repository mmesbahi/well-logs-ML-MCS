```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error

# Set visual style
sns.set_style("whitegrid")

df = pd.read_csv('../dataset/15_9-23.csv') 


print(f"Data Loaded. Shape: {df.shape}")
# --- CONFIGURATION ---
# The physics-based features we KNOW work
features = ['GR', 'RHOB', 'NPHI', 'CALI'] 
target = 'log_RDEP'

# --- DATA CLEANING ---
# 1. Select specific columns + RDEP
# We drop NaNs here to create our "Ground Truth" dataset
df_clean = df[features + ['RDEP']].dropna().copy()

# 2. Log-Transform the Target
# This handles the skewness (0.2 vs 2000 ohm.m)

df_clean['log_RDEP'] = np.log10(df_clean['RDEP'])

# 3. Simulate Missingness (The Internship Challenge)
def create_missing_data(dataframe, target_col, frac=0.2, random_state=42):
    df_missing = dataframe.copy()
    np.random.seed(random_state)
    n_samples = int(len(dataframe) * frac)
    missing_idx = np.random.choice(dataframe.index, n_samples, replace=False)
    
    true_values = dataframe.loc[missing_idx, target_col].values
    df_missing.loc[missing_idx, target_col] = np.nan
    
    return df_missing, missing_idx, true_values

# Generate the problem set
df_miss, missing_idx, y_true = create_missing_data(df_clean, target, frac=0.2)

print(f"âœ… Data Ready.")
print(f"Features: {features}")
print(f"Total Rows: {len(df_clean)} | Missing Rows to Predict: {len(y_true)}")
```

```python
# --- 1. SPLIT DATA ---
# Train on what we have, Predict on what is missing
train_df = df_miss[df_miss[target].notna()]
test_df = df_miss[df_miss[target].isna()]

X_train = train_df[features]
y_train = train_df[target]
X_test = test_df[features]

# --- 2. TRAIN MACHINE LEARNING MODEL ---
print("Training Random Forest...")
model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

# Get the "Point Prediction" (The Standard ML Result)
y_pred_mean = model.predict(X_test)

# --- 3. CALIBRATE UNCERTAINTY (The "Senior" Step) ---
print("Calibrating Uncertainty (Cross-Validation)...")
# We predict on the TRAINING data using CV to see the 'real' error
y_cv_pred = cross_val_predict(model, X_train, y_train, cv=5, n_jobs=-1)
residuals = y_train - y_cv_pred

# Calculate Sigma (Standard Deviation of the residuals)
sigma_calibrated = np.std(residuals)
print(f"âœ… Calibrated Sigma: {sigma_calibrated:.4f}")

# --- 4. MONTE CARLO SIMULATION ---
print("Running Monte Carlo Simulation...")
n_simulations = 100
safety_factor = 1  # Adjusts for non-Gaussian geological tails

mc_predictions = []

for pred in y_pred_mean:
    # Sample from Normal Distribution centered on Prediction
    # Width = Sigma * Safety Factor
    sims = np.random.normal(loc=pred, scale=sigma_calibrated * safety_factor, size=n_simulations)
    mc_predictions.append(sims)

mc_predictions = np.array(mc_predictions) # Shape: (N_missing, 100)

# Calculate 90% Confidence Intervals (5th to 95th percentile)
lower_bound = np.percentile(mc_predictions, 5, axis=1)
upper_bound = np.percentile(mc_predictions, 95, axis=1)
mc_mean_trace = np.mean(mc_predictions, axis=1) # Should match y_pred_mean closely
```

```python
# --- CORRECTED EVALUATION (ALIGNMENT FIX) ---

# 1. RETRIEVE TRUE VALUES CORRECTLY
# We use test_df.index to make sure we grab the truth in the EXACT same order as the prediction
# df_clean has the original data, test_df has the missing rows
y_true_aligned = df_clean.loc[test_df.index, target].values

# 2. CALCULATE METRICS
# A. Coverage (Reliability)
inside_band = (y_true_aligned >= lower_bound) & (y_true_aligned <= upper_bound)
coverage_score = np.mean(inside_band)

# B. RMSE Improvement (Accuracy)
# Baseline: Global average of training data
rmse_simple = np.sqrt(mean_squared_error(y_true_aligned, np.full(y_true_aligned.shape, np.mean(y_train))))

# Ours: ML Prediction
rmse_model = np.sqrt(mean_squared_error(y_true_aligned, y_pred_mean))

# Improvement Calculation
improvement = (1 - (rmse_model / rmse_simple)) * 100

print(f"--- FINAL RESULTS (ALIGNED) ---")
print(f"RMSE (Simple Average): {rmse_simple:.4f}")
print(f"RMSE (Our Model):      {rmse_model:.4f}")
print(f"ðŸš€ Improvement:        {improvement:.2f}% (Should be Positive now!)")
print(f"Uncertainty Coverage:  {coverage_score:.1%} (Target: 90%)")

# --- 3. PLOT THE RESULTS ---
plt.figure(figsize=(14, 6))

# Zoom in on a specific section
zoom_start, zoom_end = 0, 150 
indices = range(zoom_end - zoom_start)

# Plot Ground Truth (Grey)
plt.plot(indices, y_true_aligned[zoom_start:zoom_end], 
         color='black', alpha=0.5, linewidth=2, label='Ground Truth')

# Plot Mean Prediction (Blue Dashed)
plt.plot(indices, y_pred_mean[zoom_start:zoom_end], 
         color='blue', linestyle='--', linewidth=1.5, label='ML Mean Prediction')

# Plot Monte Carlo Uncertainty Band (Green)
plt.fill_between(indices, 
                 lower_bound[zoom_start:zoom_end], 
                 upper_bound[zoom_start:zoom_end], 
                 color='green', alpha=0.3, 
                 label=f'MCS 90% Confidence Interval')

plt.title("Coupling ML with Monte Carlo: Imputation with Uncertainty Quantification", fontsize=14)
plt.xlabel("Sample Index (Sorted)")
plt.ylabel("Log10(Resistivity)")
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# --- SWITCHING TO UNCALIBRATED (NAIVE) SIGMA ---

# 1. Calculate Sigma based on TRAINING data only
# The model has "seen" this data, so the error will be very small (tight bands)
y_train_pred = model.predict(X_train)
residuals_train = y_train - y_train_pred
sigma_naive = np.std(residuals_train)

print(f"Old (Calibrated) Sigma: {sigma_calibrated:.4f}")
print(f"New (Naive) Sigma:      {sigma_naive:.4f} (Much tighter!)")

# 2. RUN MONTE CARLO (No Safety Factor)
# We remove the 1.5 multiplier. We trust the model raw.
n_simulations = 100
mc_predictions_tight = []

for pred in y_pred_mean:
    # Use sigma_naive
    sims = np.random.normal(loc=pred, scale=sigma_naive, size=n_simulations)
    mc_predictions_tight.append(sims)

mc_predictions_tight = np.array(mc_predictions_tight)

# 3. CALCULATE NEW BOUNDS
lower_tight = np.percentile(mc_predictions_tight, 5, axis=1)
upper_tight = np.percentile(mc_predictions_tight, 95, axis=1)

# 4. PLOT
plt.figure(figsize=(14, 6))
zoom_start, zoom_end = 0, 150 
indices = range(zoom_end - zoom_start)

# Truth
plt.plot(indices, y_true_aligned[zoom_start:zoom_end], color='black', alpha=0.5, linewidth=2, label='Ground Truth')
# Prediction
plt.plot(indices, y_pred_mean[zoom_start:zoom_end], color='blue', linestyle='--', linewidth=1.5, label='ML Mean')
# Tight Band
plt.fill_between(indices, lower_tight[zoom_start:zoom_end], upper_tight[zoom_start:zoom_end], 
                 color='green', alpha=0.3, label='Uncalibrated 90% CI')

# Check coverage
inside_tight = (y_true_aligned >= lower_tight) & (y_true_aligned <= upper_tight)
print(f"New Coverage: {np.mean(inside_tight):.1%}")

plt.title("Coupling ML with Monte Carlo: Intrinsic Uncertainty (Tight Bands)")
plt.legend()
plt.show()
```
